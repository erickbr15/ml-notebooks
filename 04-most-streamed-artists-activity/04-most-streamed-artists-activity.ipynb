```python
import numpy as np
import pandas as pd

data = pd.read_csv('data/MostStreamedArtists.csv')
data = pd.DataFrame(data)
data.head()
```

```python
streamed_artist = data[['Lead Streams', '#']].dropna().head(10)

print(streamed_artist.dtypes)
streamed_artist.head()
```

```python
n = streamed_artist.__len__()
x = streamed_artist['Lead Streams'].to_numpy()
y = streamed_artist['#'].to_numpy()

x = x.astype(np.int64)

print(f'n={n}')
print(f'x={x}')
print(f'y={y}')
```
